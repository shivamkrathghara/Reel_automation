```python
import edge_tts
import asyncio

text = """
Most men stay average.

Not because they lack talent.

Not because they lack opportunity.

They stay average because comfort becomes their lifestyle.

The bed is comfortable.

The excuses are comfortable.

The average life is comfortable.

Growth is not.

Most people choose comfort.

A few choose discipline.

And those few change their lives.

Built in silence.
"""

async def generate_voice():

    communicate = edge_tts.Communicate(
        text=text,
        voice="en-US-BrianMultilingualNeural",
        rate="-10%"
    )

    await communicate.save("reel.mp3")

    print("✅ Voice generated successfully!")

asyncio.run(generate_voice())
```
